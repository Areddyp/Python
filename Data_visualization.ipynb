```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('churn_data.csv')

# Check column names
print("Column names in the dataset:", df.columns)

# Visualization 1: Churn Distribution (Bar Chart)
if 'Churn' in df.columns:
    churn_counts = df['Churn'].value_counts()
    plt.figure(figsize=(6, 4))
    churn_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
    plt.title('Churn Distribution')
    plt.xlabel('Churn')
    plt.ylabel('Number of Customers')
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.savefig('churn_distribution.png')  # Save the plot as an image
    plt.show()
else:
    print("Column 'Churn' not found in the dataset.")

# Visualization 2: Monthly Charges Distribution by Churn (Box Plot)
if 'MonthlyCharges' in df.columns and 'Churn' in df.columns:
    plt.figure(figsize=(8, 5))
    df.boxplot(column='MonthlyCharges', by='Churn', patch_artist=True, 
               boxprops=dict(facecolor='lightgreen', color='blue'),
               medianprops=dict(color='red'))
    plt.title('Monthly Charges Distribution by Churn')
    plt.suptitle('')  # Remove the default subtitle
    plt.xlabel('Churn')
    plt.ylabel('Monthly Charges ($)')
    plt.grid(linestyle='--', alpha=0.7)
    plt.savefig('monthly_charges_by_churn.png')  # Save the plot as an image
    plt.show()
else:
    print("Columns 'MonthlyCharges' or 'Churn' not found in the dataset.")

# Visualization 3: Tenure vs. Total Charges Colored by Churn (Scatter Plot)
if 'tenure' in df.columns and 'TotalCharges' in df.columns and 'Churn' in df.columns:
    plt.figure(figsize=(8, 5))
    colors = {'Yes': 'red', 'No': 'blue'}  # Color coding for churn
    plt.scatter(df['tenure'], df['TotalCharges'], c=df['Churn'].map(colors), alpha=0.6)
    plt.title('Tenure vs. Total Charges Colored by Churn')
    plt.xlabel('Tenure (Months)')
    plt.ylabel('Total Charges ($)')
    plt.grid(linestyle='--', alpha=0.7)
    plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='Churn: Yes'),
                         plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Churn: No')],
               title='Churn')
    plt.savefig('tenure_vs_total_charges_by_churn.png')  # Save the plot as an image
    plt.show()
else:
    print("Columns 'tenure', 'TotalCharges', or 'Churn' not found in the dataset.")
```
